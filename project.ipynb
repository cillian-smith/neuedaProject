```python
import pandas as pd
import re

# Assuming your data is in a CSV file named 'census_data.csv'
df_income_by_block_group

def parse_column_name(col_name):
    # Extract information from the column name
    match = re.match(r'Block Group (\d+); Census Tract (\d+); (.+) County; (.+)!!(Estimate)', col_name)
    if match:
        return {
            'Block Group': match.group(1),
            'Census Tract': match.group(2),
            'County': match.group(3),
            'State': match.group(4),
            'Type': match.group(5)
        }
    return None

# Create a list to store the new data
new_data = []

# Iterate through each column
for col in df_income_by_block_group.columns:
    if col != 'Label (Grouping)':  # Skip the income range column
        parsed = parse_column_name(col)
        if parsed:
            # For each row in the original dataframe
            for idx, row in df_income_by_block_group.iterrows():
                new_data.append({
                    'Block Group': parsed['Block Group'],
                    'Census Tract': parsed['Census Tract'],
                    'County': parsed['County'],
                    'State': parsed['State'],
                    'Type': parsed['Type'],
                    'Income Range': row['Label (Grouping)'],
                    'Value': row[col]
                })

# Create a new dataframe from the collected data
new_df = pd.DataFrame(new_data)

# Set the multi-index
new_df.set_index(['Block Group', 'Census Tract', 'County', 'State', 'Type', 'Income Range'], inplace=True)

# Display the first few rows of the new dataframe
print(new_df.head())

# Optionally, save the new dataframe to a CSV file
new_df.to_csv('cleaned_census_data.csv')
```

```python
def income_range_to_top(range_str):
    if 'Less than' in range_str:
        return 10000  # Midpoint of 0-10000
    elif 'or more' in range_str:
        return 250000  # Arbitrary value for $200,000 or more
    else:
        low, high = map(lambda x: x.replace('$', '').replace(',', ''),
                        range_str.split(' to '))
        return int(high)
```

```python
def calculate_county_statistics(group):
    # Convert income ranges to numeric values (top of range)
    income_tops = group.index.get_level_values('Income Range').map(income_range_to_top)
    
    # Ensure 'Value' column is numeric
    group['Value'] = pd.to_numeric(group['Value'], errors='coerce')
    
    # Calculate weighted sum and total count
    weighted_sum = (income_tops * group['Value']).sum()
    total_count = group['Value'].sum()
    
    # Calculate mean
    mean_income = weighted_sum / total_count if total_count > 0 else 0
    
    # Calculate median
    cumulative_count = 0
    median_income = 0
    sorted_group = group.sort_index(level='Income Range')
    for income, row in zip(income_tops, sorted_group.itertuples()):
        cumulative_count += row.Value
        if cumulative_count >= total_count / 2:
            median_income = income
            break
    
    return pd.Series({
        'Mean Income': mean_income,
        'Median Income': median_income,
        'Total Households': total_count
    })

# Group by county and calculate statistics
county_stats = df_income_range.groupby(level='County').apply(calculate_county_statistics)

# Display the results
print(county_stats)
```
